## Problem 1: Unique Words in a Sentence

### Problem Statement:
Given a sentence, identify all unique words and display:
- The count of unique words
- The unique words themselves

```python
# Taking input from the user
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to get unique words
unique_words = set(words)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## Problem 2: Highest Salary from Employee Data

### Problem Statement:
Store employee names and their salaries, and determine which employee has the highest salary.

```python
# Employee salary data
employees = {
     "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Get list of employee names
names = list(employees.keys())

# Assume first employee has highest salary
highest_employee = names[0]
highest_salary = employees[highest_employee]

# Compare with remaining employees
for name in names:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_employee = name

# Print final result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

## Problem 3: Find Maximum and Minimum Values

### Problem Statement:
Find the maximum and minimum values from a list of numbers.

```python
# Given list of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
max_value = numbers[0]
min_value = numbers[0]

# Loop through the list
for num in numbers:
    
    # Check for maximum
    if num > max_value:
        max_value = num
    
    # Check for minimum
    if num < min_value:
        min_value = num

# Print results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

## Problem 4: Count Products Above a Price Threshold

### Problem Statement:
Count how many products have a price greater than 1000.


```python
# Given list of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Check each price
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

## Problem 5: Calculate Attendance Percentage
### Problem Statement:
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

```python
# Given attendance list
attendance = ["P", "P", "A", "P", "P"]

# Initialize counter for present days
present_count = 0

# Count number of "P"
for day in attendance:
    if day == "P":
        present_count += 1

# Calculate total days
total_days = len(attendance)

# Calculate attendance percentage
percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

## Problem 7: Count Character Frequency

### Problem Statement:
Count how many times each character appears in a given string.

```python
# Given string
text = "python"

# Create empty dictionary
char_count = {}

# Loop through each character
for ch in text:
    
    # If character already exists, increase count
    if ch in char_count:
        char_count[ch] += 1
    else:
        # If not, add it with count 1
        char_count[ch] = 1

# Print result
print(char_count)
```

## Problem 10: Calculate Average Marks

### Problem Statement:
Calculate the average marks of a student from a list of marks.

```python
# Given list of marks
marks = [70, 75, 80, 65]

# Initialize total
total = 0

# Calculate sum of marks
for mark in marks:
    total += mark

# Calculate number of subjects
count = len(marks)

# Calculate average
average = total / count

# Print result
print("Average Marks:", average)
```
