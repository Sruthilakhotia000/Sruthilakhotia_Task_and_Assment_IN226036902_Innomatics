```python
from langchain_core.runnables import RunnableLambda
from langchain_core.prompts import PromptTemplate

print("✅ Environment Ready")
```

## 🤖 1. LLMs 

Large Language Models (LLMs) are the core engines of LangChain.

###  In this notebook:
We simulate an LLM using `RunnableLambda` to demonstrate behavior without external APIs.

###  Flow:
Input → Model → Output

```python
def fake_llm(prompt):
    return f"🤖 AI Response: {prompt}"

llm = RunnableLambda(fake_llm)

print("✅ Mock LLM Initialized")
```

```python
response = llm.invoke("What is LangChain?")
print(response)
```

## 🧩 2. Prompt Templates

Prompt templates allow dynamic input generation.

###  Why needed:
- Avoid hardcoding prompts
- Improve reusability

```python
template = PromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

prompt = template.format(topic="LangChain")
print(prompt)
```

## 🔗 3. Chains

Chains connect components into pipelines.

###  Flow:
Prompt → LLM → Output

```python
chain = template | llm

response = chain.invoke({"topic": "AI Agents"})
print(response)
```

## 🧠 4. Memory

Memory enables context retention.

###  Why needed:
- Supports conversational systems

```python
chat_history = []

def memory_llm(input_text):
    chat_history.append(input_text)
    return f"🧠 Memory: {chat_history}"

memory_chain = RunnableLambda(memory_llm)

print(memory_chain.invoke("Hello"))
print(memory_chain.invoke("What is LangChain?"))
```

## 🛠️ 5. Tools

Tools allow LLMs to perform external operations.

```python
def calculator_tool(input_text):
    try:
        return f"Result: {eval(input_text)}"
    except:
        return "Invalid expression"

tool = RunnableLambda(calculator_tool)

print(tool.invoke("5 + 10"))
```

## 🤖 6. Agents

Agents decide which tool to use dynamically.

```python
def agent(input_text):
    if "calculate" in input_text:
        return calculator_tool(input_text.replace("calculate ", ""))
    return fake_llm(input_text)

agent_chain = RunnableLambda(agent)

print(agent_chain.invoke("calculate 10 + 5"))
print(agent_chain.invoke("Explain LangChain"))
```

## 📄 7. Document Retrieval

Simulates searching relevant documents.

```python
documents = [
    "LangChain is a framework",
    "It helps build LLM applications"
]

def retrieve(query):
    return [doc for doc in documents if query.lower() in doc.lower()]

retriever = RunnableLambda(retrieve)

print(retriever.invoke("framework"))
```

## 🔍 8. Vector Search (Conceptual)

Simulates semantic similarity search.

```python
def vector_search(query):
    return f"Most relevant result for '{query}'"

vector_db = RunnableLambda(vector_search)

print(vector_db.invoke("LangChain"))
```

## 🏗️ Architecture Flow

User Input → Prompt → LLM → Chain → Agent/Tool → Output

```python
pipeline = template | llm

print(pipeline.invoke({"topic": "LangChain Architecture"}))
```
