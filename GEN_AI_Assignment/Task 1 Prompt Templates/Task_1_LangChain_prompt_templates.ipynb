## 💬 Task 4: ChatPromptTemplate System (15%)

### 🧠 Build a chat-based prompt system using roles:

- ⚙️ System → defines behavior  
- 👤 User → dynamic input  

### 🎭 Roles:
- 👨‍🏫 teacher  
- 🎤 interviewer  
- 💪 motivator  

### 🧪 Example:
role = 'teacher'  
topic = 'Neural Networks'

````python
def get_chat_prompt(role, topic):
    if role == "teacher":
        system_msg = "You are a helpful teacher."
    elif role == "interviewer":
        system_msg = "You are an interviewer asking insightful questions."
    elif role == "motivator":
        system_msg = "You are a motivational speaker."
    else:
        system_msg = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "{topic}")
    ])

    return chat_template.format_messages(topic=topic)

# ▶️ Example
messages = get_chat_prompt("teacher", "Neural Networks")
print(messages)
````

## ✅ Task 5: Input Validation Layer (10%)

### 🛠️ Create function:
validate_inputs(audience, tone)

### 📏 Rules:
- audience ∈ [beginner, intermediate, expert]  
- tone ∈ [formal, casual, fun]  

### ⚠️ If invalid:
- ❌ Raise error OR  
- 🔄 Assign default values  

````python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("⚠️ Invalid audience! Assigning default: beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("⚠️ Invalid tone! Assigning default: formal")
        tone = "formal"

    return audience, tone

# 🧪 Example
print(validate_inputs("kids", "funny"))
````

## ⚡ Task 6: Prompt Generator App (15%)

### 🛠️ Create function:
generate_prompt(topic, audience, tone, style)

### 📋 Requirements:
- ✅ Use PromptTemplate  
- 🎭 Styles: teaching, interview, storytelling  

### 🧾 Example Output:
Explain Neural Networks for beginners in a fun storytelling style

````python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = "Explain {topic} clearly step by step for {audience} in a {tone} tone"
    elif style == "interview":
        template = "Ask 3 questions about {topic} for {audience} in a {tone} tone"
    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"
    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt_template.format(topic=topic, audience=audience, tone=tone)

# ▶️ Example Output
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
````

## 🔁 Task 7: Template Reusability Test (10%)

### 📌 Instructions:
- Use ONE template  
- Run with 5 different inputs  

### 🎯 Goal:
- Same structure  
- Different outputs  

````python
# ♻️ Reusable Template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# 🧪 5 different inputs
inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "casual"),
    ("Blockchain", "expert", "formal"),
    ("Cyber Security", "beginner", "formal"),
    ("Data Science", "intermediate", "fun")
]

for topic, audience, tone in inputs:
    print(reusable_template.format(topic=topic, audience=audience, tone=tone))
````
