## Problem 1: Smart Light Controller

### Problem Statement:
Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
# Creating the SmartLight class
class SmartLight:
    
    # Constructor
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status
        
    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"
        
    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"
        
    # Method to display status
    def display_status(self):
        print(self.light_name + " is " + self.status)


# -----------------------------
# Sample Input (as given in question)
light_name = "Bedroom Light"
action = "ON"
# -----------------------------

# Creating object
light = SmartLight(light_name)

# Performing action
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()

# Displaying output
light.display_status()
```

## Problem 2: Employee ID Card System

### Problem Statement:
Create a class that stores employee name, ID, and department and displays ID card details.

```python
# Creating the Employee class
class Employee:
    
    # Constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
        
    # Method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# -----------------------------
# Sample Input (as given in question)
name = "Rahul"
emp_id = "EMP102"
department = "IT"
# -----------------------------

# Creating object
employee = Employee(name, emp_id, department)

# Displaying ID card details
employee.display_id_card()
```

## Problem 3: Mobile Contact Record

### Problem Statement:
Create a class to store contact name and phone number and display contact information.

```python
# Creating the Contact class
class Contact:
    
    # Constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
        
    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# -----------------------------
# Sample Input (as given in question)
name = "Anita"
phone = "9876543210"
# -----------------------------

# Creating object
contact = Contact(name, phone)

# Displaying contact details
contact.display_contact()
```

## Problem 4: Product Price Tag Generator

### Problem Statement:
Create a class that stores product name and price and prints a formatted price tag.

```python
# Creating the Product class
class Product:
    
    # Constructor
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    # Method to print price tag
    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: â‚¹" + str(self.price))


# -----------------------------
# Sample Input (as given in question)
name = "Headphones"
price = 2499
# -----------------------------

# Creating object
product = Product(name, price)

# Displaying price tag
product.print_price_tag()
```

## Problem 5: Movie Rating Display System

### Problem Statement:
Create a class that stores movie name and rating and displays movie details.

```python
# Creating the Movie class
class Movie:
    
    # Constructor
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
        
    # Method to display movie details
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", str(self.rating) + " / 5")


# -----------------------------
# Sample Input (as given in question)
name = "Inception"
rating = 4.8
# -----------------------------

# Creating object
movie = Movie(name, rating)

# Displaying movie details
movie.display_movie()
```

## Problem 6: Delivery Address Manager

### Problem Statement:
Create a class that stores customer name and delivery address and prints delivery details.

```python
# Creating the Delivery class
class Delivery:
    
    # Constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
        
    # Method to display delivery details
    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# -----------------------------
# Sample Input (as given in question)
customer_name = "Suman"
address = "Hyderabad"
# -----------------------------

# Creating object
delivery = Delivery(customer_name, address)

# Displaying delivery details
delivery.display_delivery()
```
