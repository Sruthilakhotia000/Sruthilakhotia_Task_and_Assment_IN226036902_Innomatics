# Problem 1: Social Media â€“ Post Engagement Analyzer

## Problem Statement
Create a function to analyze engagement on a social media post.



```python
def analyze_post_engagement(likes_list):
    
    total_likes = 0
    
    # Using loop to calculate total likes
    for like in likes_list:
        total_likes = total_likes + like
    
    # Using condition to check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example input (likes received in different hours)
likes_data = [200, 150, 300, 250, 150]

analyze_post_engagement(likes_data)
```

# Problem 3: Agriculture â€“ Rainfall Adequacy Checker
## Problem Statement
Create a function that analyzes seasonal rainfall data.



```python
def rainfall_checker(rainfall_list, required_level):
    
    total_rainfall = 0
    count = 0
    
    # Using loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall = total_rainfall + rain
        count = count + 1
    
    # Calculating average rainfall
    average_rainfall = total_rainfall / count
    
    # Using conditional logic to check adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)


# Example input (seasonal rainfall data)
seasonal_rainfall = [70, 75, 68, 80, 67]
required_rainfall_level = 70

rainfall_checker(seasonal_rainfall, required_rainfall_level)
```

# Problem 4: Social Media â€“ Duplicate Account Detection
##  Problem Statement
Create a function to detect duplicate usernames in a system.



```python
def detect_duplicate_accounts(usernames):
    
    unique_users = set()
    duplicate_found = False
    
    # Using loop and set to check duplicates
    for user in usernames:
        if user in unique_users:
            duplicate_found = True
        else:
            unique_users.add(user)
    
    # Using conditional logic
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example input
username_list = ["john123", "alice45", "john123", "mike78"]

detect_duplicate_accounts(username_list)
```

# Problem 6: Agriculture â€“ Premium Crop Price Filter
## Problem Statement
Create a function that filters crop prices above a premium threshold.



```python
def filter_premium_crops(price_list):
    
    premium_crops = []
    
    # Using loop to filter premium prices
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)


# Example input
crop_prices = [1500, 2500, 1800, 3200, 1900]

filter_premium_crops(crop_prices)
```

# Problem 9: E-Learning â€“ Student Attendance Eligibility System

## Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.

```python
def check_exam_eligibility(attendance_list):
    
    total_days = 0
    present_days = 0
    
    # Using loop to count total and present days
    for day in attendance_list:
        total_days = total_days + 1
        if day == "Present":
            present_days = present_days + 1
    
    # Calculating attendance percentage
    attendance_percentage = (present_days / total_days) * 100
    
    # Using conditional logic to check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)


# Example input
attendance_data = ["Present", "Present", "Absent", "Present", "Present"]

check_exam_eligibility(attendance_data)
```

# Problem 10: Smart Electricity Bill Analyzer
## Problem Statement
You are building a smart electricity billing system for residential users.

```python
def calculate_electricity_bill(units):
    
    bill = 0
    
    # Calculating bill based on slab system
    if units <= 100:
        bill = units * 3
    
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Classifying usage level
    if bill < 500:
        usage_status = "Low Usage"
    
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    
    else:
        usage_status = "High Usage"
    
    return bill, usage_status


# Example input
units_consumed = 250

total_bill, usage = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", usage)
```
