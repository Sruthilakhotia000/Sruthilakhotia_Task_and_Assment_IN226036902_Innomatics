## Question 1: User Login Check
Given a username and password, check whether login is successful.

```python
#predefined credentials
stored_username = "admin"
stored_password = "1234"

# Taking input from the user
username = input("Enter Username: ")
password = input("Enter Password: ")

# Checking whether both username and password match
if username == stored_username and password == stored_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## Question 2: Pass / Fail Analyzer
Analyze student results.



```python
# Given list of student marks
marks = [45, 78, 90, 33, 60]

# Initializing counters
pass_count = 0
fail_count = 0

# Checking each student's marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Printing the final result
print("Total Students :", len(marks))
print("Number of Students Passed :", pass_count)
print("Number of Students Failed :", fail_count)
```

## Question 3: Simple Data Cleaner
Clean and standardize user names.



```python
# Given list of names
names = [" Alice ", "bob", " CHARLIE "]

# Creating an empty list to store cleaned names
cleaned_names = []

# Processing each name
for name in names:
    # Remove extra spaces
    name = name.strip()
    
    # Convert to lowercase
    name = name.lower()
    
    # Add cleaned name to new list
    cleaned_names.append(name)

# Printing the final cleaned list
print("Cleaned Names List:")
print(cleaned_names)
```

## Question 4: Message Length Analyzer
Analyze message sizes.



```python
# Given list of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Checking each message
for message in messages:
    length = len(message)
    
    print("Message:", message)
    print("Length:", length)
    
    # Flag if message is longer than 10 characters
    if length > 10:
        print("Status: Long Message")
    
    print("-------------------")
```

## Question 5: Error Message Detector

Detect error messages from system logs.



```python
# Given list of logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Step 1: Initialize counter
error_count = 0

# Step 2: Count number of "ERROR" entries
for log in logs:
    if log == "ERROR":
        error_count += 1

# Step 3: Print total error count
print("Total ERROR Count:", error_count)
```
