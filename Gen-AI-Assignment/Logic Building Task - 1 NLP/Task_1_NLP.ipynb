# 📝 Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as different tokens due to case sensitivity. However, in most preprocessing steps, we convert text to lowercase to ensure uniformity and avoid duplication of meaning.

👉 Example:
- "Love" → may represent a proper noun or start of sentence
- "love" → general word

To avoid duplication, we usually convert all text to lowercase.

### 2. What happens if stopwords are not removed?
If stopwords (like "is", "the", "and") are not removed:
- They increase noise in the dataset
- Model training becomes slower
- Important words may lose significance due to frequent irrelevant words

👉 Example:
Sentence: "This is the best movie"
Tokens without removal → ["this", "is", "the", "best", "movie"]  
Tokens after removal → ["best", "movie"]

### 3. When removing stopwords can be harmful
- Sentiment Analysis: Words like "not" change meaning ("not good" ≠ "good")
- Question Answering: Words like "who", "what", "where" are essential

👉 Example 1 (Sentiment Analysis):
"I am not happy"  
Removing "not" → "I am happy" ❌ (Wrong meaning)

👉 Example 2 (Question Answering):
"What is your name?"  
Removing "what" → loses intent ❌

### 4. Stemming vs Lemmatization

👉 Example:
Word: "running"

- Stemming → "run" / "runn" (may be incorrect)
- Lemmatization → "run" (correct base form)
- 
| Feature | Stemming | Lemmatization |
|--------|--------|--------------|
| Approach | Rule-based cutting | Dictionary-based |
| Output | May not be real word | Always meaningful word |
| Example | "running" → "run" or "runn" | "running" → "run" |



# ⚙️ Task 2: Build Advanced Preprocessing Function

Implement a function that:
- Removes numbers
- Removes extra spaces
- Handles repeated characters
- Converts to lowercase
- Removes short tokens (≤2 except "no", "not")
- Removes URLs and emails

```python
import re

def preprocess_text(text):
    """
    Preprocess input text and return cleaned tokens and sentence.
    """
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs and emails
    text = re.sub(r"http\S+|www\S+|\S+@\S+", "", text)
    
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # Normalize repeated characters (loooove → love)
    text = re.sub(r"(.)\1{2,}", r"\1", text)
    
    # Remove special characters
    text = re.sub(r"[^a-z\s]", "", text)
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # Tokenization
    tokens = text.split()
    
    # Remove short tokens (≤2), keep "no" and "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]
    
    # Reconstruct cleaned sentence
    cleaned_sentence = " ".join(tokens)
    
    return tokens, cleaned_sentence
```

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for i, text in enumerate(sample_inputs, 1):
    tokens, clean = preprocess_text(text)
    
    print(f"\n🔹 Example {i}")
    print("Original Text   :", text)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", clean)
```

# 📊 Task 4: Token Analytics

For each processed sentence, compute:
1. Total number of tokens
2. Number of unique tokens
3. Average token length

```python
for text in sample_inputs:
    tokens, clean = preprocess_text(text)
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = round(sum(len(t) for t in tokens) / total_tokens, 2) if tokens else 0
    
    print("\nSentence:", text)
    print("Total Tokens :", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Length   :", avg_length)
```

# 📈 Task 5: Frequency Analysis

Combine all tokens and identify:
- Top 10 most frequent words
- Top 5 least frequent words

```python
from collections import Counter

all_tokens = []

for text in sample_inputs:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("🔝 Top 10 Frequent Words:")
print(counter.most_common(10))

print("\n🔻 Top 5 Least Frequent Words:")
print(counter.most_common()[:-6:-1])
```

# 🔗 Task 6: Build Full Pipeline

Create a function that:
- Takes a list of text inputs
- Returns:
  - Tokens list
  - Cleaned sentences

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.append(tokens)
        clean_sentences.append(clean)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Execute pipeline
output = full_pipeline(sample_inputs)
output
```
