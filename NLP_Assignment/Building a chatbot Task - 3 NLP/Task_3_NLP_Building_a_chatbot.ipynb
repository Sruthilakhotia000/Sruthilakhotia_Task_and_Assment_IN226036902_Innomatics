# 🤖 Building a Chatbot using Hugging face Transformers

## Objective
To build a console-based chatbot using a pre-trained transformer model from Hugging Face that can generate meaningful responses and simulate a basic conversational AI assistant.

---

## Technologies Used
- Python
- Hugging Face Transformers
- PyTorch
- Jupyter Notebook





### Key Steps:
1. Load pre-trained transformer model
2. Accept user input from console
3. Encode input into tokens
4. Generate response using model
5. Decode and display output
6. Continue until user types "exit" or "quit"

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch
```

## 🧠 Model Loading

We load the **flan-t5-base** model from Hugging Face.

- Pre-trained conversational model
- Handles multi-turn dialogue
- Generates human-like responses

```python
model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

## 💬 Chatbot Logic

We implement:
- Continuous conversation loop
- Context maintenance using chat history
- Exit condition (exit/quit)

```python
def run_chatbot():

    print("Chatbot: Hello! I am your AI assistant. How can I help you today?\n")

    while True:
        # ------------------------------
        # USER INPUT
        # ------------------------------
        user_input = input("You: ")

        # ------------------------------
        # EXIT CONDITION
        # ------------------------------
        if user_input.lower() in ["exit", "quit"]:
            print("Chatbot: Goodbye! Have a great day! 👋")
            break

        # ------------------------------
        # PROMPT ENGINEERING
        # ------------------------------
        prompt = (
            "You are a knowledgeable AI assistant.\n"
            "Answer clearly, accurately, and professionally.\n"
            "Keep the answer concise and well-structured.\n\n"
            f"Question: {user_input}\n"
            "Answer:"
        )

        # ------------------------------
        # TOKENIZATION
        # ------------------------------
        inputs = tokenizer(prompt, return_tensors="pt")

        # ------------------------------
        # GENERATE RESPONSE
        # ------------------------------
        outputs = model.generate(
            **inputs,
            max_length=120,
            temperature=0.3,
            top_p=0.9,
            do_sample=True
        )

        # ------------------------------
        # DECODE RESPONSE
        # ------------------------------
        response = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # ------------------------------
        # CLEAN RESPONSE
        # ------------------------------
        response = response.strip()

        if "." in response:
            response = response.split(".")[0] + "."

        if len(response) < 10:
            response = "I’m here to help. Let me provide a clear explanation."

        response = response[0].upper() + response[1:]

        # ------------------------------
        # DISPLAY OUTPUT
        # ------------------------------
        print("Chatbot:", response)


# RUN
run_chatbot()
```
